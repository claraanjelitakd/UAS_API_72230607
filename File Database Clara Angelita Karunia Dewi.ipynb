# [dbo].[ViewSalesProductWithCustomers]

```sql
/****** Object:  View [dbo].[ViewSalesProductWithCustomers]    Script Date: 13/06/2025 06:42:34 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[ViewSalesProductWithCustomers]
AS
SELECT dbo.Customers.CustomerId, dbo.Customers.CustomerName, dbo.Categories.CategoryID, dbo.Categories.CategoryName, dbo.Products.ProductId, dbo.Sales.SaleId, dbo.Sales.SaleDate, dbo.Sales.TotalAmount, dbo.SaleItems.SaleItemId, 
                  dbo.SaleItems.Quantity, dbo.Products.ProductName, dbo.Products.Price
FROM     dbo.Products INNER JOIN
                  dbo.Categories ON dbo.Products.CategoryId = dbo.Categories.CategoryID INNER JOIN
                  dbo.SaleItems ON dbo.Products.ProductId = dbo.SaleItems.ProductId INNER JOIN
                  dbo.Sales ON dbo.SaleItems.SaleId = dbo.Sales.SaleId INNER JOIN
                  dbo.Customers ON dbo.Sales.CustomerId = dbo.Customers.CustomerId
GO
```
